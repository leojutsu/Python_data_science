Algoritmo de árvore de descisão

A árvore de decisão é um algoritmo de machine learning supervisionado utilizado para classificação e regressão. Ele constroe um modelo de decisão na forma de uma árvore de nós, onde cada nó representa uma decisão ou um atributo e cada ramo representa uma possível responta ou um resultado.

Em Python podemos usar a bibliotéca Scikit-learn para implementar a árvore de decisão. A seguir, um exemplo de como utilizar o algoritmo de árvore de decisão para classificar flores no dataser iris.

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
#Carregando os dataset Iris
iris = load_iris()
```

```python
# Dividindo os dados em conjunto de treino e teste
X_train, X_test, y_train, y_test = train_test_split(iris.data,iris.target, test_size=0.2, random_state=42)
```

```python
# Criando uma instância do modelo de árvore de decisão
dtc = DecisionTreeClassifier()
```

```python
# Treinando o modelo com os dados de treino
dtc.fit(X_train, y_train)
```

```python
# Fazendo previsões com os dados de teste
y_pred = dtc.predict(X_test)
```

```python
# Avaliando a performance do modelo utilizando a acurácia
acc = accuracy_score(y_test, y_pred)
print("Acurácia:", acc)
```

```python
# Plotando a árvore de decisão
plt.figure(figsize=(20,10))
plot_tree(dtc, filled=True)
plt.show()
```

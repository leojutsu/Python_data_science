Dados Faltantes

Tratar dados nulos em ciência de dados com python pode ser feito usando diferentes técnicas, incluíndo:

1. Remoção de linhas ou colunas com dados nulos

2. Substituição por valor médio, mediana ou moda

3. Interpolação ou extrapolação de dados

4. Previsão de valores baseada em modelos de machine learning

5. Atribuição de valor zero ou outro valor constante

O método apropriado depende do contexto e da quantidade de dados nulos. É importante avaliar a influência dos dados nulos no resultado final antes de tomar uma decisão sobre como tratá-los.

Tipos de cópia de dataframe

1. Superficial
2. Profunda

Substituir dados nulos por valor médio, mediana ou moda

```python
#algum dado da coluna cm é nulo?
data['cm'].isnull().any()
```

```python
#substituir os valore nulos da coluna cm pelos valores médios (média)
data['cm'].fillna(data['cm'].mean(),inplace = True)
```

```python
#verificando para ver se os valores foram substituídos
data['cm'].isnull().any()
```

```python
data['kg'].fillna(data['kg'].median(), inplace=True)
```

```python
data['kg'].isnull().any()
```

```python
moda = data['goal'].mode().iloc[0]
data['goal'].fillna(moda, inplace=True)
data
```

```python
data['ass'].fillna(0, inplace = True)
data
```

Visualização de dados

A visualização de dados em forma de gráfico é importante porque:

1. facilita a compreessão dos dados completos

2. Permite identificar tendências e padrões mais facilmente

3. Ajuda na tomada de decisão baseada em dados

4. Torna a apresentação de dados mais clara e atrativa

5. Permite comparar informações de forma eficiente


```python
plt.scatter(x = data['kg'], y = data['cm'], label = "peso x altura")
plt.legend()
plt.xlabel('Peso (kg)')
plt.ylabel('altura (cm)')
plt.title('Peso x Altura')
plt.show()
```

```python
plt.hist(data['age'], bins = 40, label = 'Idade')
plt.legend()
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.title('Distribuição de Idade')
plt.show()
```

```python
pos = data.groupby('pos')['pos'].count()
pos
```

```python
plt.pie(pos, autopct = '%1.1f%%')
plt.title('gráfico de torta')
plt.show()
```

```python
# importando as bibliotecas necessárias
from sklearn.datasets import load_breast_cancer
import pandas as pd

# carregadores o conjunto de dados
breast_cancer = load_breast_cancer()

# criando um DataFrame com os dados do conjuntos de dados
breast_cancer_df = pd.DataFrame(breast_cancer.data, columns = breast_cancer.feature_names)

# adicionado a coluna 'target' ao DataFrame
breast_cancer_df['target'] = breast_cancer.target

# remover as linhas com valores ausentes (NaN)
breast_cancer_df.dropna()

# exibindo as primeiras linhas do DataFrame após a limpeza dos dados
print(breast_cancer_df.head())
```
